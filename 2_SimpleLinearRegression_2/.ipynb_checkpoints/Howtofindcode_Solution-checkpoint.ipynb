# Simple Linear Regression

## Importing the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

## Loading raw data

```python
#How to read data from anywhere on your drive
raw_data = pd.read_csv(r'C:\Users\alexa\E7_PC-Shop.csv')
raw_data.head()
```

# Simple Linear Regression

## Define the dependent and the independent variables

```python
y = raw_data ['Sales']
x1 = raw_data ['ShopTime']
```

## Explore the data

```python
plt.scatter(x1,y)
plt.xlabel('ShopTime', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0983*x1 - 54.4467
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line')
plt.xlabel('ShopTime', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.show()
```

## Additional Lecture: Preparing Data Sets per Brand

```python
#Find link to the description of the function below
def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]

data_ibm = filter_rows_by_values(raw_data, "Brand", ["Apple","Microsoft"])
data_apple = filter_rows_by_values(raw_data, "Brand", ["IBM","Microsoft"])
data_microsoft = filter_rows_by_values(raw_data, "Brand", ["Apple","IBM"])
```

## Run Regressions per Sub-Data File

```python
y = data_ibm ['Sales']
x1 = data_ibm ['ShopTime']
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
y = data_apple ['Sales']
x1 = data_apple ['ShopTime']
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
y = data_microsoft ['Sales']
x1 = data_microsoft ['ShopTime']
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
